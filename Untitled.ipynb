```python
import pandas as pd
from datetime import date, datetime
from random import randrange

df_new = pd.read_excel('profile.xlsx')

df1 = df_new.T

new_header = df1.iloc[0] #grab the first row for the header
df1 = df1[1:] #take the data less the header row
df1.columns = new_header #set the header row as the df header


today = date.today()
time =datetime.now()
d1= today.strftime("%m%d%y")

d2 = time.strftime("%H%M")

y = df1.loc[df1.index[0], 'DUNS_NO']
x = str(y)


i_c = randrange(999999999)
g_c = randrange(999999999)
i_control = str(i_c)

cont = df1.loc[df1.index[0],'CONTNUM']
z = str(cont)
contnum = z.replace('-','')

cat_mods=df1.loc[df1.index[0],'CAT_MODS']
ref = df1.loc[df1.index[0],'REF_FILE']
ref_file = cont+'_'+ref




data = [['ISA','00',' ','00',' ','01',x,'01','DMCOGDEN',d1,d2,'U','00305',i_control,'1','T','>'],
       ['GS','SC',contnum,'01296639',d1,d2,g_c,'X','003050'],
       ['ST','832','001'],
       ['BCT','CP',cat_mods,'','01','','','','','','05'],
       ['REF','GC',cont],
       ['REF','W6','MAS'],
       ['REF','TN', ref_file],
       ['REF','92',cat_mods],
       ['REF','ME','','V_WWW'],
       ['DTM','007','201006','','','20']
       ],


df = pd.DataFrame(data)


df[2] = df[2].replace(' ','         ')
df[4] = df[4].replace(' ','         ')


df = df.fillna('')

df2 = df.iloc[[0],[6]]
df2

def char_len(x, fixed_n):
    
    if len(x) > fixed_n: 
        return x[:fixed_n] 
    elif len(x) < fixed_n: 
#         return 'x' * (fixed_n - len(x)) + x 
        return x + (fixed_n-len(x))*' '
    return x 


c1 =df.iloc[0,6]
c2 =df.iloc[0,8]

a1 = char_len(c1,15)
a2 = char_len(c2,15)

df.iloc[[0],[6]]= a1

df.iloc[[0],[8]]= a2


# df


sep = '*'
df.applymap(str).apply(
    axis=1, func=lambda s: sep.join(el for el in s if el) 
).to_csv(
    'edi.edi', index=False, header=False
)


# df.to_csv('edi.txt',sep="*", index=False, header=False)


```
