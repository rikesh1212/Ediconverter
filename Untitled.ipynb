```python
import pandas as pd
from datetime import date, datetime
from random import randrange
import random
import string

df_new = pd.read_excel('profile.xlsx')



df1 = df_new.T

new_header = df1.iloc[0] #grab the first row for the header
df1 = df1[1:] #take the data less the header row
df1.columns = new_header #set the header row as the df header


today = date.today()
time =datetime.now()
d1= today.strftime("%m%d%y")

d2 = time.strftime("%H%M")

y = df1.loc[df1.index[0], 'DUNS_NO']
x = str(y)


i_c = randrange(999999999)
g_c = randrange(999999999)
i_control = str(i_c)

cont = df1.loc[df1.index[0],'CONTNUM']
z = str(cont)
contnum = z.replace('-','')

cat_mods=df1.loc[df1.index[0],'CAT_MODS']
ref = df1.loc[df1.index[0],'REF_FILE']
ref_file = cont+'_'+ref

mo = df1.loc[df1.index[0],'MIN_ORD']
mop = str(mo)+'00'

web = df1.loc[df1.index[0],'V_WWW']
# sin = df1.loc[df1.index[0],'SIN']
prmpt_disc = df1.loc[df1.index[0],'PRMPT_DISC']
prmpt_days = df1.loc[df1.index[0],'PRMPT_DAYS']
c_deliv = df1.loc[df1.index[0],'C_DELIV']
sin = df1.loc[df1.index[0],'SIN1']
vend_name = df1.loc[df1.index[0],'VENDNAME']
v_str1 = df1.loc[df1.index[0],'V_STR1']
v_city = df1.loc[df1.index[0],'V_CITY']
v_state = df1.loc[df1.index[0],'V_STATE']
v_zip = df1.loc[df1.index[0],'V_ZIP']
v_ctry = df1.loc[df1.index[0],'V_CTRY']
v_phone = df1.loc[df1.index[0],'V_PHONE']
v_fax = df1.loc[df1.index[0],'V_FAX']
v_email = df1.loc[df1.index[0],'V_EMAIL']
a_name = df1.loc[df1.index[0],'A_NAME']
a_phone = df1.loc[df1.index[0],'A_PHONE']
a_fax = df1.loc[df1.index[0],'A_FAX']
a_email = df1.loc[df1.index[0],'A_EMAIL']

fob_hi = df1.loc[df1.index[0],'FOB_HI']
next_fob_hi = fob_hi+'A'

fob_ak = df1.loc[df1.index[0],'FOB_AK']
next_fob_ak = fob_ak+'A'

fob_pr = df1.loc[df1.index[0],'FOB_PR']
next_fob_pr = fob_pr+'A'


stc = string.ascii_uppercase+string.digits
stc_control = ''.join(random.choice(stc) for i in range(8))



data = [['ISA','00',' ','00',' ','01',x,'01','012966339',d1,d2,'U','03050',i_control,'1','T','>'],
       ['GS','SC',x,'012966339',d1,d2,g_c,'X','003050'],
       ['ST','832',stc_control],
       ['BCT','CP',cat_mods,'','01','','','','','','05'],
       ['REF','GC',cont],
       ['REF','W6','MAS'],
       ['REF','TN','',ref_file],
       ['REF','92',cat_mods],
       ['REF','ME','',web],
       ['DTM','007','201006','','20'],
       ['CTB','OR','','','','MO',mop],
       ['CTB','OR',sin,'','','MA','25000000'],
       ['ITD','','',prmpt_disc,'',prmpt_days,'','','','','','','','','N'],
       ['LDT','AF',c_deliv,'DA'],
       ['FOB','PP','OR'],
       ['N1','40','','1','012966339'],
       ['N1','SE',vend_name,'1',x],
       ['N3',v_str1],
       ['N4',v_city,v_state,v_zip,v_ctry],
       ['REF','4A'],
       ['PER','OD',vend_name,'TE',v_phone,'FX',v_fax,'EM',v_email],
       ['PER','AC',a_name,'TE',a_phone,'FX',a_fax,'EM',a_email],
       ['N1','DZ','','ZN','00'],
       ['N4','','HI','','',next_fob_hi],
       ['N1','DZ','','ZN','00'],
       ['N4','','AK','','',next_fob_ak],
       ['N1','DZ','','ZN','00'],
       ['N4','','PR','','',next_fob_pr],
       ]

       
    


df = pd.DataFrame(data)


df[2] = df[2].replace(' ','          ')
df[4] = df[4].replace(' ','          ')



df = df.fillna('')

df2 = df.iloc[[0],[6]]
df2

def char_len(x, fixed_n):
    
    if len(x) > fixed_n: 
        return x[:fixed_n] 
    elif len(x) < fixed_n: 
#         return 'x' * (fixed_n - len(x)) + x 
        return x + (fixed_n-len(x))*' '
    return x 


c1 =df.iloc[0,6]
c2 =df.iloc[0,8]

a1 = char_len(c1,15)
a2 = char_len(c2,15)

df.iloc[[0],[6]]= a1

df.iloc[[0],[8]]= a2

mod_i_df = pd.read_excel('IPROD.xlsx')

second_df = pd.read_excel('IPHOTO.xlsx')

mod_ii_df = pd.merge(mod_i_df,second_df[['CONTNUM','DEF_PHOTO']], on='CONTNUM', how='left')

i_df = mod_ii_df.reindex(['N1','DZ','ZN','00','N4','HI','FOB_HI','AK','FOB_AK','PR','FOB_PR','LIN','1','MG','MFGPART','ZZ',
                     'MFGNAME','PD','PRODNAME','UP','UPC','CH','PPOINT','PI','SIN','VP','VENDPART','REF','ME','','P_WWW','51',
                 'DEF_PHOTO','PID','F','DESCRIPTION','MEA','DE','TT','WARNUMBER','WARPERIOD','PO4','UNT','WEIGHT','LB','D_VOLUME','D_UNIT',
                    'VALUE1','VALUE2','VALUE3','UNIT1','LDT','LEADTIME','P_DELIV','DA'],axis=1)
i_df['LIN'] = i_df['LIN'].fillna('LIN')

m = i_df['1'].isna()
i_df.loc[m,'1'] = range(1, m.sum() + 1)

i_df['MG'] = i_df['MG'].fillna('MG')
i_df['ZZ'] = i_df['ZZ'].fillna('ZZ')
i_df['PD'] = i_df['PD'].fillna('PD')
i_df['UP'] = i_df['UP'].fillna('UP')
i_df['CH'] = i_df['CH'].fillna('CH')
i_df['PI'] = i_df['PI'].fillna('PI')
i_df['VP'] = i_df['VP'].fillna('VP')
i_df['N1'] = i_df['N1'].fillna('N1')
i_df['DZ'] = i_df['DZ'].fillna('DZ')
i_df['ZN'] = i_df['ZN'].fillna('ZN')
i_df['00'] = i_df['00'].fillna('00')
i_df['N4'] = i_df['N4'].fillna('N4')
i_df['HI'] = i_df['HI'].fillna('HI')
i_df['AK'] = i_df['AK'].fillna('AK')
i_df['PR'] = i_df['PR'].fillna('PR')
i_df['FOB_HI'] = i_df['FOB_HI'].astype(str)+'A'
i_df['FOB_PR'] = i_df['FOB_PR'].astype(str)+'A'
i_df['FOB_AK'] = i_df['FOB_AK'].astype(str)+'A'
i_df['REF'] = i_df['REF'].fillna('REF')
i_df['ME'] = i_df['ME'].fillna('ME')
i_df[''] = i_df[''].fillna('')
i_df['51'] = i_df['51'].fillna('51')
i_df['DEF_PHOTO'] = i_df['DEF_PHOTO'].fillna('')
i_df['PID'] = i_df['PID'].fillna('PID')
i_df['F'] = i_df['F'].fillna('F')

i_df['DESCRIPTION'] = mod_i_df['PRODDESC']

i_df['MEA'] = i_df['MEA'].fillna('MEA')
i_df['DE'] = i_df['DE'].fillna('DE')
i_df['TT'] = i_df['TT'].fillna('TT')

i_df['WARPERIOD'] = i_df['WARPERIOD'].astype(str).replace('YEAR','YR')
i_df['WARPERIOD'] = i_df['WARPERIOD'].astype(str).replace('MONTH','MM')
i_df['WARPERIOD'] = i_df['WARPERIOD'].astype(str).replace('DAYS','DA')

i_df['PO4'] = i_df['PO4'].fillna('PO4')
i_df['UNT'] = i_df['UNT'].fillna('UNT')
i_df['LB'] = i_df['LB'].fillna('LB')
i_df['LDT'] = i_df['LDT'].fillna('LDT')
i_df['DA'] = i_df['DA'].fillna('DA')

# f =  lambda data: [data[x:x+80] for x in range(0, len(data), 80)]
# s = i_df['DESCRIPTION'].apply(f).explode().reset_index(drop=True)
# i_df['DESCRIPTION'] = s
# i_df['DESCRIPTION'] = i_df['DESCRIPTION'].fillna('')


vals = [['N1','DZ','ZN','00'],['N4','','HI','','','FOB_HI'],['N1','DZ','ZN','00'],['N4','','AK','','','FOB_AK'],
       ['N1','DZ','ZN','00'],['N4','','PR','','','FOB_PR'],
       ['LIN','1','MG','MFGPART','ZZ','MFGNAME','PD','PRODNAME','UP','UPC','CH','PPOINT','PI','SIN','VP','VENDPART'],
       ['REF','ME','','P_WWW'],
       ['REF','51','DEF_PHOTO'],
       ['PID','F','','','','DESCRIPTION'],
       ['MEA','DE','TT','WARNUMBER','WARPERIOD'],
       ['PO4','','','','UNT','','WEIGHT','LB','D_VOLUME','D_UNIT','VALUE1','VALUE2','VALUE3','UNIT1'],
       ['LDT','LEADTIME','P_DELIV','DA']]


L = [i_df.loc[:, x].set_axis(range(len(x)), axis=1) for x in vals]
i_df = pd.concat(L).sort_index(kind='mergesort').fillna('').reset_index(drop=True)
i_df = i_df.fillna('')

frames = [df,i_df]
result = pd.concat(frames)

result = result.fillna('')




sep = '*'
result.applymap(str).apply(
    axis=1, func=lambda s: sep.join(s).strip('*')
).to_csv(
    'edi.edi', index=False, header=False
)



# with open("edi.edi", "w") as f: 
#     f.write("".join([l.rstrip("*") for l in df.to_csv(sep="*",header=None, index=None).split("\n")]))

# # # df.to_csv('edi.txt',sep="*", index=False, header=False)

# with open("edi.edi") as f: print(f.read())
```
